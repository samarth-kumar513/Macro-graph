```python
df3 = pd.read_excel("C:/Users/vtvt4/Desktop/fig1_v.xlsx")
sns.set(rc={'figure.figsize':(13,7)})
sns.set_style("darkgrid")
x1 = sns.regplot(x="Log SGDP per capita (2000)", y="Log consumption per capita (2000)",data=df3, ci = None, line_kws={'color': 'red'},scatter_kws={"color":"darkred","alpha":0.3,"s":200} );
for line in range(0,df3.shape[0]):
         x1.text(df3["Log SGDP per capita (2000)"][line]+0.01, df3["Log consumption per capita (2000)"][line], 
                 df3["state"][line], horizontalalignment='left', 
                 size='large', color='black', weight='bold')
```
